```python
import matplotlib.pyplot as plt
import numpy as np
```

matplotlib travaille essentiellement avec la bibliotheque numpy.
par exemple on utilisera Les fonctions  np.linspace(), np.arange() et les fonctions du  module *random* de Numpy pour generer les cordonnées (x,y) pour le tracer des graphes 

```python
#exemple : tracer de la fonction x^2
x=np.linspace(0,2,10)
y=x**2 # pas possible dans sans numpy ( brodcasting)
print(x)
plt.plot(x,y)
```

### 1) Pyplot-Style

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,10,20)
x=x.reshape((20,1))
y=x + np.random.randn(20,1)
#nuage de points
plt.scatter(x,y)


#ajouter des styles ( Nom de la courbe , epaisseur du trait, style du trait ,couleur dur trait)
#possibilite infinie
#c => color / lw => line with (epaisseur)/ ls => line style (style de ligne),.....
plt.plot(x,y,c='red',lw=8,ls='',label='Courbe')

#dir(plt)
```

Astuce pour pour tracer d'autre graphe : taper plt. puis Tab

on peut ajouter un titre , changer le style des points , changer le noms des axes , changer l'epaisseur du trac, ajouter une legende etc... on a énorment de possibilité
https://matplotlib.org/3.3.2/tutorials/introductory/customizing.html

Normalement pour creer une figure on doit commencer par pyplot.figure() et terminé par pyplot.close() ( cycle de vie d'une figure)

plt.show() sert à afficher le graphe mais on a pas besoin de lui dans jupyter
plt.savefig() permet de sauvegarder l'image

```python
#le cycle de vie d'une figure
plt.figure()
plt.plot(x,y,label='quadratique')
plt.plot(x,x+1,label='Cubique')
plt.title('figure 1')
plt.xlabel('axe X')
plt.ylabel('axe Y')
plt.legend()
plt.show()
plt.savefig('figuer1.png')
```

```python
#plusieurs graphique sur un meme graphique (subplot sans s !!!)
plt.subplot(2,1,1)
plt.plot(x,y,c='red')
plt.title('graphique 1')
plt.subplot(2,1,2)
plt.plot(x,y,c='green')
plt.title('graphique 2')
```

### 2) OO-Style

```python
#creer des graphiques avec la meme ordonnee
#ax est un tableau numpy 
fig,ax= plt.subplots(2,1) 
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show() #pas utile dans jupyter
```

```python
type(ax) #tableau numpy
```

Avec le parametre *sharex* on peut avoir deux figures qui parle le meme axe des abscisses.

```python
fig,ax= plt.subplots(2,1,sharex=True) 
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
```
